## Exercise 4 - SQLite

1 Add the necessary libraries for SQLite3 and make a database named `test.db`.

```python
import sqlite3
import pandas as pd

db_conn = sqlite3.connect("test.db", timeout=10)
```

2 Create a connection to the database.

```python
c = db_conn.cursor()

# c.execute('DROP TABLE IF EXISTS people')
```

3 Make the following table in the database, named `people`:
    
| name | age | shoe_size |
| ------------- |-------------| -----|
| Prank | 88 | 47 |
| Matt | 34 | 39 |
| Ilona | 41 | 36 |
| Sonja | 32 | 37 |



```python
c.execute('''
CREATE TABLE IF NOT EXISTS people (
    name TEXT,
    age INTEGER,
    shoe_size INTEGER
)
''')

people_data = [
    ('Prank', 88, 47),
    ('Matt', 34, 39),
    ('Ilona', 41, 36),
    ('Sonja', 32, 37)
]

c.executemany('INSERT INTO people VALUES (:name, :age, :shoe_size)', people_data)

db_conn.commit()
```

4 When the Table has been created, convert the table to the Pandas DataFrame object and add the column `hat_size' to it, then fill that column with the values ​​you choose.

```python
df = pd.read_sql_query('SELECT * FROM people', db_conn)
df['hat_size'] = [58, 57, 56, 55]

print(df)
```

5 Make a copy of the DataFrame object with the name `person_table_copy` and add a row to it with the values ​you choose. Now save this modified dataframe object in the database as `person_table_copy` (will be created next to the previous table).

```python
person_table_copy = df.copy()
new_row = {'name': 'Anna', 'age': 21, 'shoe_size': 40, 'hat_size': 55}
person_table_copy = pd.concat([person_table_copy, pd.DataFrame([new_row])], ignore_index=True)

person_table_copy.to_sql('person_table_copy', db_conn, if_exists='replace', index=False)

print(person_table_copy)
```

6 Finally, save the database changes and close the connection.

```python
db_conn.commit()
db_conn.close()
```
